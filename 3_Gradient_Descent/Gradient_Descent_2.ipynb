```python
def grad_descent(x,y):
    m=b=0
    learning_rate = 0.001
    n = len(x)
    prev_cost =0
    
    for i in range(1000):
        
        y_predicted = m*x+b
        print('prediction for y by current formula', y_predicted)
        print('actual y', y)
        cost = (1/n) * sum([val**2 for val in (y-y_predicted)])
        #update m and b by implementing partial derivatives
        
        md = -(2/n) * sum(-x*(y - (m*x +b)))
        bd = (2/n) * sum(-y+(m*x+b))
        #can also be written as
        #md=-(2/n) * sum(x*(y-y_predicted))
        #bd=(2/n) * sum(-y+y_predicted)
        
        #update
        m = m-learning_rate * md
        b = b -learning_rate +bd
        
        
        if math.isclose(cost, prev_cost, rel_tol=1e-20):
            break
        prev_cost = cost
        print(f'm {m}, cost {cost}, b {b}, iteration {i}')
        #update to improve cost
```

```python
grad_descent(x,y)
```
